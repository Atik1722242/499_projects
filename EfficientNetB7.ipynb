```python
DATA_DIR = '/content/AugDataset'
```

```python
def find_desired_set(x):
    return x.split('/')[3]
```

```python
image_paths = glob.glob(DATA_DIR + '/*/*/*.*')

images = []
classes = []

for img in image_paths:
    images.append(img)
    classes.append(img.split('/')[-2])

df_dict = {
    'paths' : images,
    'classes' : classes
}

df = pd.DataFrame(df_dict)

dummies = pd.get_dummies(df['classes'])
df = pd.concat([df, dummies], axis = 1)

df['set'] = df['paths'].apply(find_desired_set)

df
```

```python
train_df = df.loc[df['set'] == 'Train'].iloc[:, :-1]
valid_df = df.loc[df['set'] == 'Valid'].iloc[:, :-1]
```

```python
import torch
import torchvision
```

```python
class LoadDataset(torch.utils.data.Dataset):
    def __init__(self, df, transform=None):
        self.df = df
        self.transform = transform
    
    def __len__(self):
        return len(self.df)
    
    def __getitem__(self, idx):
        img_path = self.df.iloc[idx, 0]
        image = cv2.imread(img_path)
        if self.transform:
            image = self.transform(image)
        label = torch.tensor(self.df.iloc[idx, 2:])
        return (image, label)
```

```python
transform = torchvision.transforms.Compose([
    torchvision.transforms.ToPILImage(),
    torchvision.transforms.Resize((128, 128)),
    # torchvision.transforms.RandomAffine(degrees=(90, 180), translate=(0.1, 0.3), scale=(0.5, 0.75)),
    torchvision.transforms.ToTensor()
])
```

```python
dataset_train = LoadDataset(df=train_df, transform=transform)
dataset_valid = LoadDataset(df=valid_df, transform=transform)

train_loader = torch.utils.data.DataLoader(dataset=dataset_train, batch_size=32, shuffle=True, drop_last=False, pin_memory=True)
valid_loader = torch.utils.data.DataLoader(dataset=dataset_valid, batch_size=32, shuffle=False, drop_last=False, pin_memory=True)
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
device
```

```python
model = torchvision.models.efficientnet_b7(pretrained=True)

model.classifier[1] = torch.nn.Linear(
    in_features=model.classifier[1].in_features,
    out_features= len(set(list(df.iloc[:, 1]))) # 13
)
```

```python
model.to(device)

optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
loss_fn = torch.nn.CrossEntropyLoss()
```

```python
def save_model(model, train_loss, epoch, PATH):
    torch.save({
        'epoch': epoch,
        'model_state_dict': model.state_dict(),
        # 'optimizer_state_dict': optimizer.state_dict(),
        'loss': train_loss
    }, PATH)
    print(f"---------\nModel Saved at {PATH}\n---------\n")
```

```python
def load_model(model, PATH):
    checkpoint = torch.load(PATH)
    model.load_state_dict(checkpoint['model_state_dict'])
    # optimizer.load_state_dict(checkpoint['optimizer_state_dict'])
    epoch = checkpoint['epoch']
    train_loss = checkpoint['loss']
    return checkpoint, epoch, train_loss
```

```python
best_acc = 0
starting_epoch = 0
N_EPOCHS = 100
PATH = '/content/drive/MyDrive/PROJECTS/TreeClassifier/Checkpoint/efficientnet_b7.pth'

if os.path.exists(PATH):
    ckp, starting_epoch, train_loss = load_model(model, PATH)

for epoch in range(starting_epoch, N_EPOCHS-1):
    model.train()
    train_losses = []
    train_count, correct_preds = 0, 0
    train_loss = 0.
    for i, (images, labels) in enumerate(tqdm(train_loader)):
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        optimizer.zero_grad()
        loss = loss_fn(outputs.to(float), labels.to(float))
        loss.backward()
        optimizer.step()
        _, preds = torch.max(outputs.data, 1)
        _, labels = torch.max(labels.data, 1)
        train_count += labels.size(0)
        correct_preds += (preds == labels).sum().item()
        train_loss += loss.item() * labels.size(0)
        # train_losses.append(loss.item())
    train_acc = (correct_preds / train_count) * 100.
    train_loss = train_loss / train_count


    # MODEL EVALUATION
    model.eval()
    valid_count, correct_preds_valid = 0, 0
    with torch.no_grad():
        for i, (images, labels) in enumerate(valid_loader):
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, preds = torch.max(outputs.data, 1)
            _, labels = torch.max(labels.data, 1)
            valid_count += labels.size(0)
            correct_preds_valid += (preds == labels).sum().item()

    valid_acc = (correct_preds_valid / valid_count) * 100.

    if epoch > 0 and valid_acc > best_acc:
        best_acc = valid_acc
        save_model(
            model, train_loss, epoch, PATH
        )

    print(f'Epoch: {epoch}, Train Loss: {train_loss:.2f}, Train Acc: {train_acc:.2f}%, Valid Acc: {valid_acc:.2f}%')
```

```python
actual_labels, predictions = [], []

ckp, starting_epoch, train_loss = load_model(model, PATH)
model.eval()
with torch.no_grad():
    for i, (images, labels) in enumerate(valid_loader):
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        _, preds = torch.max(outputs.data, 1)
        _, labels = torch.max(labels.data, 1)
        labels = labels.detach().cpu().numpy()
        preds = preds.detach().cpu().numpy()
        for i in range(len(labels)):
            actual_labels.append(labels[i])
            predictions.append(preds[i])
```

```python
pd.DataFrame({
    'Prediction': predictions, 'Target': actual_labels
})
```

```python
from sklearn import metrics

PR = metrics.precision_score(actual_labels, predictions, average='macro')
RE = metrics.recall_score(actual_labels, predictions, average='macro')
F1 = metrics.f1_score(actual_labels, predictions, average='macro')
ACC = metrics.accuracy_score(actual_labels, predictions)

print(f"{'-'*17}\nPrecision: {PR:.4f}\n{'-'*17}")
print(f"   Recall: {RE:.4f}\n{'-'*17}")
print(f" F1 Score: {F1:.4f}\n{'-'*17}")
print(f" Accuracy: {ACC*100:.2f}%\n{'-'*17}")
```
